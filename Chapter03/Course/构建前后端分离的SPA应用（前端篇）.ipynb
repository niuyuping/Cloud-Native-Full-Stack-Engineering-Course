## 目标

</div>  

* 将第1章完成的<span class="terminal-span">保险计算器</span>重构为前后端分离的现代化SPA应用。我们会将保险费率、税率等敏感数据和算法分离到后端服务中，前端服务则仅专注于交互功能的实现。

* 我们将以第1章中<span class="terminal-span">Google AI Studio</span>制作的原型为基础，构建我们的第一个<span class="terminal-span">前端服务</span>。

* #### 扩展知识
  
    * <span class="terminal-span">SPA</span> 是一个整个生命周期只加载一个 HTML 文件的 <span class="terminal-span">Web</span> 应用。后续的所有交互都靠 JavaScript 动态更新页面。它解决了传统网页跳转时“刷页、白屏”的糟糕体验，让网页用起来像 App 一样流畅。
    
    * <span class="terminal-span">React</span> 就是用来构建 SPA 的核心工具。 如果没有 React 这种框架来高效地管理页面如何局部更新，手动写原生 JS 去实现 SPA 将会极其复杂和痛苦。
    
    * <span class="terminal-span">Next.js</span> 的出现很大程度上是为了解决纯 SPA 的缺点（首屏慢、SEO 差），它通过<span class="terminal-span">服务器端渲染 (SSR)</span>技术，让应用既有 SPA 的流畅体验，又有 MPA 的首屏速度和 SEO 优势。
