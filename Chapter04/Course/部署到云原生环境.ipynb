* #### 扩展知识
  
    * <span class="terminal-span">Cloud Run</span> 是在 Google Cloud 上部署无状态应用的服务，它结合了 Kubernetes 的灵活性和 Serverless 的便捷性。如果你有一个可以通 过 HTTP 访问的 Web 应用或 API，且不需要持久化存储在本地磁盘，那么 Cloud Run 以及同类服务通常是性价比最高且最省心的选择。
    
    * <span class="terminal-span">Cloud SQL</span> 是一项全托管关系型数据库服务，它在云端提供 MySQL、PostgreSQL 或 SQL Server 数据库，不需要自己去购买服务器、安装数据库软件、处理备份或打补丁，是传统 Web 应用迁移上云或新开发应用最常用的数据库选择。

## 创建数据库

</div>  

* #### 创建Cloud SQL PostgreSQL实例

    * ① 导航到[Cloud SQL开始页面](https://console.cloud.google.com/sql/get-started)。
  
    * ② 依次点击 <span class="button-div">创建实例</span> -> <span class="button-div">选择 PostgreSQL</span> 
  
    * ③ 在「创建PostgreSQL」页面中依次配置以下参数: 

        * 「Cloud SQL版本」 ->  <span class="radio-div">企业版</span>

        * 「版本预设」 ->  沙盒

        * 「实例信息」
        
            * 「数据库版本」 -> PostgreSQL 17

            * 「实例ID」 -> <span style="color: orange">_social-insurance-database_</span>

            * 「密码」 -> {**输入要使用的默认管理员密码**}

        * 「区域」 -> {**选择离你的用户最近的区域**}

        * 「可用区级可用性」　-> <span class="radio-div">单个可用区</span>

        * 「自定义实例」

            * 「机器配置」 -> 通用 - 共享核心 -> <span class="radio-div">1vCPU, 0.614 GB</span>

            * 「存储空间 -> 存储类型」 -> <span class="radio-div">HDD</span>
            
            * 「存储空间容量」 -> <span class="radio-div">10 GB</span>
        
        * 「连接数」

            * 「实例IP分配」 -> <span class="check-div">专用IP</span>

            * 「选择专用连接方法」 -> <span class="check-div">专用服务访问通道(PSA)</span> -> {**选择默认的<span class="terminal-span">VPC网络</span>**}

            * <span class="uncheck-div">公共IP</span>
     
    * ④ 点击 <span class="button-div">保存</span>
    
    <div style="margin-bottom: 1rem"></div>

* #### 创建数据库
  
    * ① 导航到[Cloud SQL实例页面](https://console.cloud.google.com/sql/instances)。
  
    * ② 依次点击「实例 -> social-insurance-database」

    * ③ 依次点击「数据库」 -> <span class="button-div">创建数据库</span>

        * 「数据库名称」 -> <span style="color: orange">_social_insurance_</span>

        * 点击 <span class="button-div">创建</span>
  
    <div style="margin-bottom: 1rem"></div>

* #### 创建数据库用户

    * ① 导航到[Cloud SQL实例页面](https://console.cloud.google.com/sql/instances)。
  
    * ② 依次点击「实例 -> social-insurance-database」

    * ③ 依次点击「用户」 -> <span class="button-div">添加用户帐号</span>

        * <span class="radio-div">内置身份验证</span>

        * 「用户名」 -> <span style="color: orange">_db_user_</span>

        * 「密码」 -> {**输入db_user的密码**}

        * 点击 <span class="button-div">添加</span>

    <div style="margin-bottom: 1rem"></div>

* #### 配置内部专用网络
  
    * ① 导航到[Cloud SQL实例页面](https://console.cloud.google.com/sql/instances)。
  
    * ② 依次点击「实例 -> social-insurance-database」

    * ③ 依次点击「连接 -> 摘要」

    * ④ 记下「专用IP连接 -> 服务连接方法 -> 内部IP地址」，在部署<span class="terminal-span">后端服务</span>时会用到这个IP地址。
  
    <div style="margin-bottom: 1rem"></div>

* #### 扩展知识
  
    * 本章建立的Cloud SQL实例是按照<span class="terminal-span">最经济的标准</span>创建的，意味着性能很低，花费也很低，主要用于学习目的。如果是生产环境，则需要按照实际需求重新配置。

* #### 创建无服务器VPC连接器

    * ① 导航到[Cloud VPC网络页面](https://console.cloud.google.com/networking)。
  
    * ② 依次点击 「无服务器VPC访问通道」 -> <span class="button-div">创建连接器</span> 按钮。
  
    * ③ 在「创建连接器」页面中依次配置以下参数: 

        * 「名称」 -> <span style="color: orange">_social-insurance-vpc_</span>

        * 「区域」 -> {**选择离你的用户最近的区域**}

        * 「网络」　-> {**选择默认网络**}

        * 「子网」 -> 「自定义IP范围」 -> 「IP范围」 -> <span style="color: orange">_10.8.0.0_</span>
        
        * <span class="button-div">显示扩缩设置</span> -> 「调节设置」

            * 「实例类型」 -> f1-micro
     
    * ④ 点击 <span class="button-div">创建</span> 按钮。
    
    <div style="margin-bottom: 1rem"></div>

* #### 扩展知识
  
    * <span class="terminal-span">VPC</span>是虚拟私有云（Virtual Private Cloud）的缩写，它构建在Google Cloud、AWS这类公有云上，<span class="terminal-span">VPC网络</span>是在VPC内部建立的专用网络，数据和流量在逻辑上与其他用户的流量完全分开，互不干扰，我们可以在VPC网络上定义独立路由和安全策略。VPC也可以通过<span class="terminal-span">虚拟专用网络（VPN）</span>或专线与<span class="terminal-span">私有云（Private Cloud）</span>连接，构建<span class="terminal-span">混合云（Hybrid Cloud）</span>。

* #### 在Cloud Run上部署后端服务

    * ① 导航到[Cloud Run服务页面](https://console.cloud.google.com/run/services)。
  
    * ② 点击 <span class="button-div">部署容器</span> 按钮。
  
    * ③ 在「创建服务」页面中依次配置以下参数: 

        * <span class="radio-div">从代码库（源代码或函数）持续部署</span> -> 点击 <span class="button-div">设置Cloud Build</span>

            * 「源代码库 -> 代码库」-> {**选择后端服务代码库**}
            
            * 点击 <span class="button-div">下一步</span>

            * 「构建配置 -> 构建类型 -> <span class="radio-div">Dockerfile</span> -> 来源位置」 -> <span style="color: orange">_/Dockerfile.prod</span>
            
            * 点击 <span class="button-div">保存</span>

        * 「“服务”名称」 ->  <span style="color: orange">_social-insurance-backend-service_</span>

        * 「区域」 -> {**选择离你的用户最近的区域**}
        
        * 「身份验证」 -> <span class="radio-div">允许公开访问</span>

        * 「容器、卷、网络、安全性 -> 容器」

            * 「变量和密钥」 -> <span class="button-div">添加变量</span> -> {**添加以下变量**}

                * <span style="color: orange">_DB_URL_</span> = <span style="color: orange">_r2dbc:postgresql://{**数据库专用IP**}:5432/social_insurance_</span>

                * <span style="color: orange">_FLYWAY_URL_</span> = <span style="color: orange">_jdbc:postgresql://{**数据库专用IP**}:5432/social_insurance_</span>

                * <span style="color: orange">_DB_USER_</span> = <span style="color: orange">_db_user_</span>
        
                * <span style="color: orange">_DB_PASSWORD_</span> = {**db_user的密码**}

          ※ 提示： 这里的{**数据库专用IP**}是指上一小节「配置内部专用网络 -> ④」中保存下来的数据库内部IP地址。

        * 「容器、卷、网络、安全性 -> 网络」

            *  <span class="check-div">连接到VPC以获取出站流量</span> 
            
            *  <span class="radio-div">使用无服务器VPC访问通道连接器</span> -> {**选择上一小节创建的VPC连接器**}

            *  <span class="radio-div">将所有流量路由到VPC</span>

    * ④ 点击 <span class="button-div">创建</span> ，等待服务创建完成。

    * ⑤ 在「服务详情」页面中找到「网址」并将其复制下来，这是<span class="terminal-span">后端服务的URL</span>，接下来在部署<span class="terminal-span">前端服务</span>时会用到。
    
    <div style="margin-bottom: 1rem"></div>

    * <span class="terminal-span">环境变量</span>（Environment Variables）是指操作系统层面的全局变量，可以被传递给运行在该系统上的应用程序，在云原生（Cloud Native）架构中，环境变量不仅仅是参数，而是「配置管理」的标准接口。12-Factor App 方法论中明确定义：“在环境中存储配置”，为<span class="terminal-span">不可变基础设施</span>（Immutable Infrastructure）提供可变的注入，实现了构建产物与部署环境的物理隔离。

    * <span class="terminal-span">跨域资源共享</span>（CORS） 是一种基于 HTTP 头的浏览器安全机制，它允许服务器明确声明“哪些外部网站有权访问我的资源”，从而有条件地打破浏览器默认的严格同源策略（Same-Origin Policy）限制。。

* #### 扩展知识
    
    * <span class="terminal-span">构建时环境变量</span>（Build-time environment variables）是指在编译、打包过程中被读取的环境变量，而<span class="terminal-span">运行时环境变量</span>（Run-time environment variables）是指在应用程序在实际启动运行时被读取的环境变量。这是非常核心的 DevOps 概念，尤其是在使用 Docker 容器（如 Cloud Run）和现代前端框架（如 Next.js）时，理解两者的区别至关重要。

    * <span class="terminal-span">DevOps</span> 是一组文化理念、实践方法和工具的集合，旨在缩短系统开发周期，同时提供高质量的持续交付。可以简单理解为<span class="terminal-span">谁开发（Dev）谁运维（Ops）</span>，开发人员应事先而不是事后提供充分的运维支持。DevOps是一种文化，强调自动化、可观测性和可持续性。

    * 顾名思义，<span class="terminal-span">服务端渲染</span>（SSR）模式是在指服务器上完成页面渲染的模式，而<span class="terminal-span">客户端渲染</span>（CSR）模式是指在客户端设备上完成页面渲染的模式，两者在用户体验、SEO优化、资源占用方式等方面有着显著差异，在现代前端架构的最佳实践中，经常根据实际需求，将两种模式组合在一起使用，形成<span class="terminal-span">混合渲染模式</span>。
