* #### 扩展知识

  * <span class="terminal-span">React</span> 是由 Meta (Facebook) 开发并维护的一个用于<span class="terminal-span">构建用户界面</span>（UI）的 JavaScript 库。它是目前全球最流行的前端开发工具之一。React 在内存中维护一个<span class="terminal-span">虚拟的页面结构</span>。当数据变化时，它会先在虚拟层计算出最小的变化量，然后再更新真实的浏览器页面。这大大提高了性能。

  * <span class="terminal-span">Next.js</span> 是一个基于 React 的现代 Web 全栈框架，它最核心的价值在于提供了开箱即用的<span class="terminal-span">服务端渲染</span>（SSR）和静态站点生成（SSG）能力，完美解决了传统 React 单页应用 SEO（搜索引擎优化）差和首屏加载慢的痛点。配合其<span class="terminal-span">基于文件系统的自动路由</span>、内置的后端 API 功能以及针对图像和字体的自动优化组件，Next.js 让开发者能够以最高效的方式构建出高性能、生产级的 Web 应用。

<div style="margin-bottom: 1rem"></div>

* #### 扩展知识
      
    * <span class="terminal-span">DevOps</span> 是 "Development"（开发）和 "Operations"（运维）两个词的组合。它不是一个职位，也不是一款软件，而是一种文化转变和工作方式，可以直观理解为 <span class="terminal-span">谁开发谁运维</span>。

    * <span class="terminal-span">CI/CD</span> 是持续集成（Continuous Integration）和持续交付/部署（Continuous Delivery/Deployment）的缩写。它是 DevOps 的一套具体的<span class="terminal-span">技术实践和流程</span>，旨在通过自动化的方式，将代码从开发者的电脑交付到生产环境。
